```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.optimize import curve_fit
from numpy import sqrt, diag
```

```python
#reading data
mantid_df = pd.read_csv('data/cut_sE150T1p6_E_102_dE_10.dat')
mantid_df = mantid_df.dropna()
```

```python
from modules.fitQ import fitQdistrib
```

```python
popt_102, pcov_102, Q_arr_102, S_arr_102, S_err_102, Q_ex_102, S_ex_102, Q_fit_102, res_102, S_err_fit_102 = fitQdistrib(mantid_df, 4.26, 12, 6.83, 7.24 )
```
